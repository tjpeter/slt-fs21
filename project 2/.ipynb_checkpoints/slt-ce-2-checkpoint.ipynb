<ul style="background-color:#F9E9C6;"> 
In [1] they refer to <i>Mass-Constrained Clustering</i> as the preferred implementation of the DA clustering algorithm.

    
This is due to fact that the usual annealing process depends on the number of coincident codevectors in each effective cluster, which is undesirable. <i>Mass-Constrained Clustering</i> therefore only keeps as many codevectors as necessary (i.e. equal to the number of clusters) and assigns them a weight (mass). The number of code vectors or centroids is then only increased, once a critical temperature - which is computable - has been reached. As such this variant of DA clustering becomes computationally more efficient.

During the usual cooling process, i.e. lowering the temperature stepwise to (almost) zero, optimal centroids are determined by an EM-like algorithm.
<ul/>

<li> Before we can compute the confusion matrix, we need to perform some post-processing on the DAC cluster assignments.
    Explain what the function postprocess (defined below) does, and why we need it. To do so, complete the docstring of the function postprocess.
        </li>

```python
def postprocess(y_DAC, y_kmeans):
    """
    Based on the resulting assignment probabilites of the DAC algorithm, 
    the most probable assignment is determined ("hard assignment"). 
    Then clusters are aligned with k-means result in order to make it comparable resp.
    usable for the confusion matrix below.
    """
    
    y_DAC_hard = np.argmax(y_DAC, axis=1)
    
    n_clusters = len(np.unique(y_DAC_hard))
    dac2kmeans = []
    for cluster in range(n_clusters):
        argmax = np.argmax(y_DAC[:, cluster])
        dac2kmeans.append(y_kmeans[argmax])
        
    y_DAC_new = []
    for dac_label in y_DAC_hard:
        y_DAC_new.append(dac2kmeans[dac_label])
        
    return np.array(y_DAC_new)
```

```python
skl.metrics.confusion_matrix(y_kmeans, postprocess(y_DAC, y_kmeans))
```

<ul style="background-color:#F9E9C6;"> 
Some of the attributes possess significantly different scales
despite having the same unit of measurement (e.g. g/dm3), and others have even different unit of measurements (e.g. alcohol % vol or pH).
Either way, using a euclidean metric assumes (implicitly) that the different attributes are on a similar scale or that distances between datapoints are meaningful resp.
comparable. Yet, a comparsion of distance/difference in fixed acidity is not comparable to a distance in pH-values. As a consequence, a 'scale-free' metric, which depends only on the ratio between measurements, might be more suited for this dataset.
